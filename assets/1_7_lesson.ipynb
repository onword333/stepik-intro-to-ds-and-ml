# Визуализация

```python
import pandas as pd
import numpy as np

#%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

students = pd.read_csv('StudentsPerformance.csv')
salary = pd.read_csv('data/income.csv')
```

Гистрограмма

```python
# задает пределы
#students.plot(ylim=(0,1000))

students['math score'].hist()
```

Точечная диаграмма (график корреляции между двумя переменными)
Результат math score и reading score: линейная взаимосвязь переменных

```python
students.plot.scatter(x = 'math score', y = 'reading score')
```

Seaborn, точечная диаграмма
Показывает регриссионную прямую, которая показывает линейную аппроксимацию взаимосвязи двух переменных

```python
# hue - группирующая переменная
# fit_reg - убирает регрессионые прямые
ax = sns.lmplot(x = 'math score', y = 'reading score', data=students, hue='gender', fit_reg=False)

# кастомизация графика
ax.set_xlabels('My math score')
```
